```python
def return_result_dataframe(command):
    connection.reconnect()
    
    cursor = connection.cursor()
    cursor.execute(command)
    data=cursor.fetchall()
    
    df = pd.DataFrame(data)
    column=[i[0] for i in cursor.description]
    df.columns=column
    
    return df
```

```python
#Looking at total cases vs Total Deaths
```

```python
command="""select location,date,total_cases,total_deaths,(total_deaths/total_cases)*100 as deathpercases
from covid_death
where continent is not null
order by 1,2"""
```

```python
df=return_result_dataframe(command)
df=df.fillna(0)
df
```

```python
x_values=[]
y_values=[]
pre_loc=df.loc[0,'location']

plt.figure(figsize=(30, 10))

for location,d,deathpercases in zip(df['location'],df['date'],df['deathpercases']):
    if(location!=pre_loc):
        
        plt.plot(x_values,y_values,label=pre_loc)
        x_values=[]
        y_values=[]
        pre_loc=location
        
    x_values.append(d.to_pydatetime().date())
    y_values.append(deathpercases)
    
plt.plot(x_values,y_values,label=pre_loc)

plt.ylabel("Death Ratio")
plt.title("Death ratio across all countries during covid")
plt.show()
```

```python
y_values=list(df['deathpercases'])

plt.figure(figsize=(30, 10))
plt.ylabel("Death Ratio")
plt.title("Death ratio graphs in india during covid")
plt.plot(x_values,y_values)
```

```python
#A pie chart visualising percentage of people in india are total vaccinated
```

```python
command="""
select distinct location, max(people_fully_vaccinated) over (PARTITION by location) as people_fully_vaccinated ,T.population
from covid_vacination,(select population from covid_death where location like '%india%') as T
where continent is not null
and location like '%india%'
"""
```

```python
df=return_result_dataframe(command)
df
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.axis('equal')
langs = ['Vaccinated', 'Not Vaccinated']
students = df.loc[0,'people_fully_vaccinated'],df.loc[0,'population']-df.loc[0,'people_fully_vaccinated']
myexplode = [0.2, 0]
ax.pie(students, labels = langs,autopct='%1.2f%%',explode=myexplode)
plt.show()
```
